```python
def find_pop_pct(df):
    tot_pop = df['POP2010'].sum()
    pop_pct = df['POP2010'] / tot_pop
    
    return pop_pct

def agg_pct(df):
    pp_df = find_pop_pct(df)
    return df[pct_columns].mul(pp_df, 0)
    
#def agg_bool():
```
