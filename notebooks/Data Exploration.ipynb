```python
pct_df = food_data[['State', 'County', 'POP2010'] + pct_columns]
bool_df = food_data[['State', 'County', 'POP2010'] + bool_columns]
```

```python
def find_pop_pct(df):
    tot_pop = df['POP2010'].sum()
    pop_pct = df['POP2010'] / tot_pop
    
    return pop_pct

def agg_pct(df):
    #print (df.columns)
    pp_df = find_pop_pct(df)
    return df[pct_columns].mul(pp_df, 0).sum()
    
def agg_bool(df):
    pp_df = find_pop_pct(df)
    return df[bool_columns].mul(pp_df, 0).sum()

def agg_income(df):
    pp_df = find_pop_pct(df)
    return df[['MedianFamilyIncome']].mul(pp_df, 0).sum()
```

```python
total_agg = food_data[['State', 'County'] + tot_columns].groupby(['State', 'County']).sum().reset_index()
pct_agg = pct_df.groupby(['State', 'County']).apply(lambda x: agg_pct(x)).reset_index()
bool_agg = bool_df.groupby(['State', 'County']).apply(lambda x: agg_bool(x)).reset_index()
income_agg = food_data[['State', 'County', 'POP2010', 'MedianFamilyIncome']].groupby(['State', 'County']).apply(lambda x: agg_income(x)).reset_index()
```
